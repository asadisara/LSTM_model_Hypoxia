```python
    
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from numpy.polynomial.polynomial import Polynomial
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split, KFold
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
from tensorflow.keras.regularizers import l1_l2
from tensorflow.keras.losses import Huber
from tensorflow.keras.optimizers import Adam
#from keras_tuner import RandomSearch
import matplotlib.pyplot as plt
import shap
import pkg_resources
import joblib
```

```python
# Load and preprocess the data using the full path
data = pd.read_csv('F:/0 UCAM/project5_SWAT+WETMachineLearning/inputdata_meteo.csv')  # Replace with the actual path to your data file
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Verify the data
#print(data.head())
print("Data head:\n", data.head())
```

```python
# Prepare the input and output columns
X_columns = ['U10.m_s', 'V10.m_s', 'SurfPressure.hPa', 'AirTemp.C', 'DewpointTemp.C', 'CloudCover_fraction', 'ShortWave.W_m2', 'Precip.m_day', 'watertemp_degC', 'density_kg/m3', 'ammonium_gN/m3', 'nitrate_gN/m3', 'phosphate_gP/m3', 'macrophyte-dryweight_gDW/m2', 'secchi-depth_m', 'totalN_gN/m3', 'totalP_gP/m3', 'total-chlorophyll-a_mg/m3','flo_in', 'sed_in', 'orgn_in', 'sedp_in', 'no3_in', 'solp_in', 'nh3_in']
y_column = 'o2sat_%_WET_6.5'
```

```python
# Scale the features using all available data
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_scaled = scaler_X.fit_transform(data[X_columns].values)
y_scaled = scaler_y.fit_transform(data[y_column].values.reshape(-1, 1)).flatten()

print("Scaled X shape:", X_scaled.shape)
print("Scaled y shape:", y_scaled.shape)

# Save the scalers
joblib.dump(scaler_X, 'F:/0 UCAM/project5_SWAT+WETMachineLearning/scaler_X.pkl')
joblib.dump(scaler_y, 'F:/0 UCAM/project5_SWAT+WETMachineLearning/scaler_y.pkl')
```

```python
# Split the data into train, validation, and test sets
X_train_val, X_test, y_train_val, y_test = train_test_split(X_scaled, y_scaled, test_size=0.15, shuffle=False)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.2941, shuffle=False) # 0.2941 is approximately 25% of 0.85


print("X_train_val shape:", X_train_val.shape)
print("X_train shape:", X_train.shape)
print("X_val shape:", X_val.shape)
print("X_test shape:", X_test.shape)
```

```python
# Define the time step
time_step = 7  # Number of time steps
```

```python
# Helper function to create sequences
def create_sequences(X, y, time_step):
    X_seq, y_seq = [], []
    for i in range(len(X) - time_step):
        X_seq.append(X[i:(i + time_step), :])
        y_seq.append(y[i + time_step])
    return np.array(X_seq), np.array(y_seq)
    
```

```python

# Create sequences for training, validation, and test sets
X_train_seq, y_train_seq = create_sequences(X_train, y_train, time_step)
X_val_seq, y_val_seq = create_sequences(X_val, y_val, time_step)
X_test_seq, y_test_seq = create_sequences(X_test, y_test, time_step)

print("X_train_seq shape:", X_train_seq.shape)
print("y_train_seq shape:", y_train_seq.shape)
print("X_val_seq shape:", X_val_seq.shape)
print("y_val_seq shape:", y_val_seq.shape)
print("X_test_seq shape:", X_test_seq.shape)
print("y_test_seq shape:", y_test_seq.shape)
```

```python
# Define the K-Fold cross-validator
k = 5  # Number of folds
kf = KFold(n_splits=k)

# Initialize variables to store metrics for each fold
fold_metrics = []
best_model = None
best_val_loss = float('inf')

# Initialize lists to store loss values
train_losses = []
val_losses = []
```

```python


# Perform k-fold cross-validation
for fold, (train_index, val_index) in enumerate(kf.split(X_train_seq)):
    print(f"Training on fold {fold+1}...")
    X_train_fold, X_val_fold = X_train_seq[train_index], X_train_seq[val_index]
    y_train_fold, y_val_fold = y_train_seq[train_index], y_train_seq[val_index]

    # Build the LSTM model with dropout
    model = Sequential()
    #model.add(LSTM(32, input_shape=(time_step, X_train_fold.shape[2])))
    model.add(LSTM(16, return_sequences=True, input_shape=(time_step, X_train_fold.shape[2]), kernel_regularizer=l1_l2(l1=0.001, l2=0.001)))
    model.add(Dropout(0.3))  # Add dropout layer with 0.2 dropout rate
    #model.add(LSTM(64, return_sequences=True))  # Second LSTM layer with return_sequences=True
    #model.add(Dropout(0.2))  # Add another dropout layer
    #model.add(LSTM(64, return_sequences=True))  # Second LSTM layer with return_sequences=True
    #model.add(Dropout(0.2))  # Add another dropout layer
    model.add(LSTM(16, kernel_regularizer=l1_l2(l1=0.001, l2=0.001)))
    model.add(Dropout(0.3))  # Add another dropout layer
    model.add(Dense(1))
    #model.compile(loss='mean_squared_error', optimizer=optimizer)
    #model.compile(loss=smape_loss, optimizer='adam')

    # Define the optimizer with a specific learning rate
    optimizer = Adam(learning_rate=0.001)
    model.compile(loss=Huber(delta=1.0), optimizer=optimizer)

    # Define the callbacks
    early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.001)
    
    # Train the model and store the history
    history = model.fit(X_train_fold, y_train_fold, epochs=100, batch_size=16, verbose=0, shuffle=True, validation_data=(X_val_fold, y_val_fold), callbacks=[early_stopping, reduce_lr])

    # Store the loss values
    train_losses.append(history.history['loss'])
    val_losses.append(history.history['val_loss'])
    
    # Predict using the trained model
    y_pred_train_fold = model.predict(X_train_fold).flatten()
    y_pred_val_fold = model.predict(X_val_fold).flatten()

    # Inverse transform to get original scale
    y_train_fold = scaler_y.inverse_transform(y_train_fold.reshape(-1, 1)).flatten()
    y_val_fold = scaler_y.inverse_transform(y_val_fold.reshape(-1, 1)).flatten()
    y_pred_train_fold = scaler_y.inverse_transform(y_pred_train_fold.reshape(-1, 1)).flatten()
    y_pred_val_fold = scaler_y.inverse_transform(y_pred_val_fold.reshape(-1, 1)).flatten()

    # Calculate metrics for the train period of the current fold
    train_metrics = {
        'NSE': 1 - (sum((y_train_fold - y_pred_train_fold)**2) / sum((y_train_fold - y_train_fold.mean())**2)),
        'RMSE': np.sqrt(mean_squared_error(y_train_fold, y_pred_train_fold)),
        'MAE': np.mean(np.abs(y_train_fold - y_pred_train_fold)),
        'PBIAS': 100 * sum(y_train_fold - y_pred_train_fold) / sum(y_train_fold)
        }

    # Calculate metrics for the validation period of the current fold
    val_metrics = {
        'NSE': 1 - (sum((y_val_fold - y_pred_val_fold)**2) / sum((y_val_fold - y_val_fold.mean())**2)),
        'RMSE': np.sqrt(mean_squared_error(y_val_fold, y_pred_val_fold)),
        'MAE': np.mean(np.abs(y_val_fold - y_pred_val_fold)),
        'PBIAS': 100 * sum(y_val_fold - y_pred_val_fold) / sum(y_val_fold)
        }

    # Append validation metrics to the fold metrics list
    fold_metrics.append({
        'Fold': fold+1,
        'Train_NSE': train_metrics['NSE'],
        'Train_RMSE': train_metrics['RMSE'],
        'Train_MAE': train_metrics['MAE'],
        'Train_PBIAS': train_metrics['PBIAS'],
        'Val_NSE': val_metrics['NSE'],
        'Val_RMSE': val_metrics['RMSE'],
        'Val_MAE': val_metrics['MAE'],
        'Val_PBIAS': val_metrics['PBIAS']
        # Add other metrics as needed
        })
    
    # Calculate validation loss
    val_loss = mean_squared_error(y_val_fold, y_pred_val_fold)

    # Save the model if it has the best validation loss so far
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_model = model
        best_model_fold = fold + 1
    
# Save the best model
best_model.save('F:/0 UCAM/project5_SWAT+WETMachineLearning/best_lstm_model.h5')

print("Fold number", best_model_fold, "is the best model")

# Convert the list of metrics to a DataFrame
fold_metrics_df = pd.DataFrame(fold_metrics)
print(fold_metrics_df)

# Save the fold metrics to a CSV file
fold_metrics_df.to_csv('F:/0 UCAM/project5_SWAT+WETMachineLearning/5-fold_4-unit_metrics_60_25_15_lstm_O2depth6.5GOTMWET.csv', index=False)
```

```python
# Split the data for final model training and testing (70%-30% split)
X_train_final, X_test_final, y_train_final, y_test_final = train_test_split(X_scaled, y_scaled, test_size=0.3, shuffle=False)

# Create sequences for the final model training and testing
X_train_final_seq, y_train_final_seq = create_sequences(X_train_final, y_train_final, time_step)
X_test_final_seq, y_test_final_seq = create_sequences(X_test_final, y_test_final, time_step)


# Build the final LSTM model with dropout
final_model = Sequential()
final_model.add(LSTM(50, return_sequences=True, input_shape=(time_step, X_train_final_seq.shape[2]), kernel_regularizer=l1_l2(l1=0.001, l2=0.001)))
final_model.add(Dropout(0.5))  # Add dropout layer with 0.2 dropout rate
#final_model.add(LSTM(64, return_sequences=True, kernel_regularizer=l1_l2(l1=0.001, l2=0.001)))  # Second LSTM layer with return_sequences=True
#final_model.add(Dropout(0.2))  # Add another dropout layer
final_model.add(LSTM(50, kernel_regularizer=l1_l2(l1=0.001, l2=0.001)))  # Third LSTM layer
final_model.add(Dropout(0.5))  # Add another dropout layer
final_model.add(Dense(1))  

# Define the optimizer with a specific learning rate
optimizer = Adam(learning_rate=0.0005)
final_model.compile(loss='mean_squared_error', optimizer=optimizer)

#final_model.compile(loss=Huber(delta=1.0), optimizer=optimizer)

# Define Early Stopping
early_stopping = EarlyStopping(monitor='loss', patience=20, restore_best_weights=True)

#Train the model
#final_model.fit(X_train_final_seq, y_train_final_seq, epochs=100, batch_size=16, verbose=0, shuffle=True)

# Train the model with Early Stopping
final_model.fit(X_train_final_seq, y_train_final_seq, epochs=100, batch_size=16, verbose=0, shuffle=True, callbacks=[early_stopping])

# Save the trained model
final_model.save('F:/0 UCAM/project5_SWAT+WETMachineLearning/final_model.h5')

# Predict for train set
y_pred_train = final_model.predict(X_train_final_seq).flatten()

# Inverse transform to get original scale
y_train = scaler_y.inverse_transform(y_train_final_seq.reshape(-1, 1)).flatten()
y_pred_train = scaler_y.inverse_transform(y_pred_train.reshape(-1, 1)).flatten()

# Calculate metrics for the train set
train_metrics = {
    'NSE': 1 - (sum((y_train - y_pred_train)**2) / sum((y_train - y_train.mean())**2)),
    'RMSE': np.sqrt(mean_squared_error(y_train, y_pred_train)),
    'MAE': np.mean(np.abs(y_train - y_pred_train)),
    'PBIAS': 100 * sum(y_pred_train - y_train) / sum(y_train)
}

# Print and save the test metrics
print('Train Metrics:', train_metrics)

# Convert the test metrics to a DataFrame
train_metrics_df = pd.DataFrame([train_metrics], index=['Train'])

# Save the test metrics to a CSV file
train_metrics_df.to_csv('F:/0 UCAM/project5_SWAT+WETMachineLearning/train_metrics_lstm_O2depth6.5GOTMWET.csv', index=True)

# Predict using the final trained model
y_pred_test = final_model.predict(X_test_final_seq).flatten()

# Inverse transform to get original scale
y_test = scaler_y.inverse_transform(y_test_final_seq.reshape(-1, 1)).flatten()
y_pred_test = scaler_y.inverse_transform(y_pred_test.reshape(-1, 1)).flatten()

# Calculate metrics for the test set
test_metrics = {
    'NSE': 1 - (sum((y_test - y_pred_test)**2) / sum((y_test - y_test.mean())**2)),
    'RMSE': np.sqrt(mean_squared_error(y_test, y_pred_test)),
    'MAE': np.mean(np.abs(y_test - y_pred_test)),
    'PBIAS': 100 * sum(y_pred_test - y_test) / sum(y_test)
}

# Print and save the test metrics
print('Test Metrics:', test_metrics)

# Convert the test metrics to a DataFrame
test_metrics_df = pd.DataFrame([test_metrics], index=['Test'])

# Save the test metrics to a CSV file
test_metrics_df.to_csv('F:/0 UCAM/project5_SWAT+WETMachineLearning/test_metrics_lstm_O2depth6.5GOTMWET.csv', index=True)

# Predict using the final trained model for the entire dataset
X_scaled_seq, _ = create_sequences(X_scaled, y_scaled, time_step)
y_pred_full = final_model.predict(X_scaled_seq).flatten()

# Inverse transform to get original scale for the full dataset predictions
y_pred_full_original_scale = scaler_y.inverse_transform(y_pred_full.reshape(-1, 1)).flatten()

# Create a DataFrame with the date index, observed and predicted values
results_df = pd.DataFrame({
    'Date': data.index[time_step:],  # Adjust the index to match the sequence length
    'ObservedQ': scaler_y.inverse_transform(y_scaled[time_step:].reshape(-1, 1)).flatten(),
    'PredictedQ': y_pred_full_original_scale
}).set_index('Date')

# Save the results to a CSV file
results_df.to_csv('F:/0 UCAM/project5_SWAT+WETMachineLearning/20250403_predicted_values_full_time_period_7-time-steps_input_O2depth6.5GOTMWET.csv', index=True)

print("y_pred_train shape:", y_pred_train.shape)
print("y_pred_test shape:", y_pred_test.shape)
```
